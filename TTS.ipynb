```python
tts = TTS(model_name="tts_models/multilingual/multi-dataset/your_tts", progress_bar=False, gpu=True)
text = "A week after Foxconn cleared its intention about splitting its semiconductor joint venture with Vedanta, Anil Agarwal-led oil-to-gas conglomerate said it is committed to building India’s first semiconductor and display fabrication plant. In this regard, Vedanta claims to have made significant progress in closing tie-ups for technology as well as equity partners for semiconductors. While the company did not disclose any details, it is expected to make an announcement soon. "

tts.tts_to_file(text, speaker_wav="sample_data/tar.wav", language="en", file_path="output.wav")
```
